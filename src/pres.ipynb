```python
import BingChatAccess
import ChatGPTAccess
import json
import os
```

```python
chatGPT = ChatGPTAccess.ChatGPTAccess()
```

```python
bingChat = BingChatAccess.BingChatAccess()
```

```python
options = {"len" : 20, "adj" : "geographical", "top" : "Central asia", "pointsMin" : 4, "pointsMax" : 6}
```

```python
with open('prompts.json') as f:
        prompts = json.load(f)
```

```python
first = prompts["topics"].format(options["len"], options["adj"], options["top"])
```

```python
answer = chatGPT.answerPropt(first, 500, newConv=True)
```

```python
print(answer)
```

```python
topics  = json.loads(answer)
topics = topics[list(topics.keys())[0]]
print(topics)
```

```python
rets = []
first = True
for top in topics:
    if first:
        question = prompts["bingTopic"].format(options["adj"], options["top"], top)
        ret = await bingChat.answerPropt(question, 700, newConv=True)
        first = False
    else:
        try:
            question = prompts["bingContinue"].format(top)
            ret = await bingChat.answerPropt(question, 700)
        except:
            question = prompts["bingTopic"].format(options["adj"], options["top"], top)
            ret = await bingChat.answerPropt(question, 700, newConv=True)
    
    print(question)
    print(ret)
    rets.append(ret)
```

```python
with open("meta.txt", "w", encoding="UTF-8") as f:
    for i in range(len(topics)):
        f.write(topics[i] + "\n")
        f.write(rets[i]["text"] + "\n\n\n")
```

```python
info = []

for i in range(len(topics)):
    info.append(topics[i])
    info.append(rets[i]["text"] + "\n\n")

text = "\n".join(info)
```

```python
with open("meta.txt", "r", encoding="UTF-8") as f:
    text = f.read()
print(text)
```

```python
with open('prompts.json') as f:
    prompts = json.load(f)

prompt = prompts["final"].format(text, options["adj"], options["top"])
print(prompt)

pres = chatGPT.answerPropt(prompt, 1000, newConv=True, temp=0.2)
print(pres)

name = "pres {}.md"
num = 1
while os.path.exists(name.format(num)):
    num+=1

with open(name.format(num), "w", encoding="UTF-8") as f:
    f.write(pres)
```
